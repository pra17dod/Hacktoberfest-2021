```python
import networkx as nx
import numpy as np


def pageRank(g,nodes,alpha,limit):
    
    dampingFactor = alpha
    pi = np.array([1/nodes]*nodes,dtype=np.float64)
    graph = np.zeros((nodes,nodes),dtype=np.float64)
    
    inCount = [0.0]*nodes
    nodesList = list(g.nodes)


    for (start,end,weight) in g.edges(data=True):
        inCount[nodesList.index(start)] += float(weight["value"])

    for (start,end,weight) in g.edges(data=True):
        graph[nodesList.index(end)][nodesList.index(start)] = float(weight["value"])/inCount[nodesList.index(start)]
    
    for i,incoming in enumerate(inCount):
        if(incoming == 0):
            graph[:,i] = [1/nodes]*nodes

    graph = dampingFactor * graph + (1-dampingFactor) * ((1/nodes) * np.ones((nodes,nodes)))

    present = pi
    prev = pi
    count = 0
    
    while(count < limit and sum(inCount) != 0):
        prev = present        
        present = np.matmul(graph,present)
        if(prev == present).all():
            break
        count += 1

    print(nodesList[list(present).index(max(present))])


g = nx.read_gml("netscience.gml",label="label")
g = g.to_directed()

pageRank(g,len(list(g.nodes())),0.85,100)
```
